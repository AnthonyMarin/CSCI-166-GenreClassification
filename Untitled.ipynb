# Training of First Model: All Genres(Speechiness, Loudness, Danceability)
Our first model with be trained off the speechiness, loudness, and danceability of songs from all genres.

Hypothesis: Due to there being some genres very similar to one another, our first model will likely have a difficult time predicting some of the rock based genres. 

```python
comboD = pnda.read_csv("dataSets/trainingData1/all_genres_SLP.csv")
```

```python
scaler = StandardScaler()
scaler.fit(comboD.drop('target',axis=1))
scaled_features = scaler.transform(comboD.drop('target',axis = 1))

all_genresSLD_feat = pnda.DataFrame(scaled_features, columns = comboD.columns[:-1])
```

```python
from sklearn.model_selection import train_test_split

x = all_genresSLD_feat
y = comboD['target']

x_train,x_test,y_train,y_test  = train_test_split(x,y,test_size=0.3, random_state = 30, shuffle = True)

```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 7, p = 2, metric = 'euclidean')
knn.fit(x_train,y_train)
```

```python
prediction = knn.predict(x_test)
prediction
```

# Results of First Model
    This model learned from all genre tables using the features speechiness, loudness, and danceability. 


```python
from sklearn.metrics import classification_report
print(classification_report(y_test, prediction))
```

Tags(1 = alt, 2 = blues, 3 = hiphop, 4 = indie, 5 = metal, 6 = pop, 7 = rock)

The results above show that our model was more likely to correctly identify a hiphop song more than any other genre. 

Metal, pop, and rock were all identified ~%40 of the time.

Blues and indie music were identified ~20-25% of the time.

Alternative was only identified 6% of the time.

For our next model, we plan to eliminate genres that might share some musical features in hopes of acquiring more accurate results. 

# Training Second Model: Hiphop/Rock(Speechiness, Loudness, Danceability)
For the second iteration of training, we will be aiming for success. 

Looking back at the first model's results, we see that rock and rap had the highest precision and recall rates. We will use these two genres in our next model. 

```python
hh_rock_SLD = pnda.read_csv("dataSets/trainingData2/hh_rock.csv")
scaler2 = StandardScaler()
scaler2.fit(hh_rock_SLD.drop('target',axis=1))
scaled_features2 = scaler2.transform(hh_rock_SLD.drop('target',axis = 1))

hh_rock_SLD_features = pnda.DataFrame(scaled_features2, columns = hh_rock_SLD.columns[:-1])

x2 = hh_rock_SLD_features
y2 = hh_rock_SLD['target']

x_train2,x_test2,y_train2,y_test2  = train_test_split(x2,y2,test_size=0.3, random_state = 30, shuffle = True)

knn2 = KNeighborsClassifier(n_neighbors = 2, p = 2, metric = 'euclidean')
knn2.fit(x_train2,y_train2)
prediction2 = knn2.predict(x_test2)
prediction2

```

# Results of Second Model

```python
print(classification_report(y_test2, prediction2))
```

# Training Third Model: Blues, Hiphop, Rock (Speechiness, Loudness, Danceability)

```python
blues_hh_rock_SLD = pnda.read_csv("dataSets/trainingData3/blues_hh_rock.csv")
scaler3 = StandardScaler()
scaler3.fit(blues_hh_rock_SLD.drop('target',axis=1))
scaled_features3 = scaler3.transform(blues_hh_rock_SLD.drop('target',axis = 1))

blues_hh_rock_SLD_features = pnda.DataFrame(scaled_features3, columns = blues_hh_rock_SLD.columns[:-1])

x3 = blues_hh_rock_SLD_features
y3 = blues_hh_rock_SLD['target']

x_train3,x_test3,y_train3,y_test3  = train_test_split(x3,y3,test_size=0.3, random_state = 30, shuffle = True)

knn3 = KNeighborsClassifier(n_neighbors = 3, p = 2, metric = 'euclidean')
knn3.fit(x_train3,y_train3)
prediction3 = knn3.predict(x_test3)
prediction3
```

```python
print(classification_report(y_test3, prediction3))
```
